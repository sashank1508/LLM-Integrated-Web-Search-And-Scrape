```python
import requests
def scrape_jina_ai(url: str) -> str:
    """Scrapes content from the Jina AI redirection URL and returns the response text."""
    response = requests.get("https://r.jina.ai/" + url)
    return response.text
```

```python
scrape_jina_ai('https://www.thehindu.com/news/national/mamata-thanks-everyone-who-supported-her-to-lead-india-bloc/article68974205.ece')
```
